```python
#Solution 5

"""
>Create a scatter plot between top two projections explaining most of variance in the digits data.
>The scatter plot is plot between top two PCs for each class label. There are total of 10 labels (0 to 9).

>Use for loop to iterate on each labelled subset digits data - 
"""

for i in range(0, len(data['label'].unique())):
    data['label'].unique()
    
    '''create subset for each class label'''
    subsetdata =  data[data['label'] == data['label'].unique()[i]]  
    label = subsetdata['label']
    subsetvar = subsetdata.drop(["label"], axis =1) # drop the label column since we will apply PCA 
    
    '''apply standard scaler to scale the data with mean 0 and unit variance'''
    scaled = StandardScaler(copy=True, with_mean=True).fit_transform(subsetvar)  # shape is (4684, 784)
    
    '''Compute eigenvalues,eigen vectors from covariance matrix of the scaled data'''
    cvm = numpy.cov(scaled, rowvar = False, bias = False )
    eigenva, eigenve = numpy.linalg.eig(cvm) # input to linalg.eig() from numpy module is square array which is covariance matrix in our case
    pairs_eigva_eigve = [(numpy.abs(eigenva[k]), eigenve[:,k]) for k in range(len(eigenva))] 
    
    '''Sort eigen values in descending order in list '''
    sorting_eig_pairs =sorted(pairs_eigva_eigve, key=lambda eigva: eigva[0], reverse=True)  
    
    '''Stack horizontally top two eigen value vector which captures maximum variance in data'''
    stackedcomp = numpy.hstack((sorting_eig_pairs[0][1].reshape(784,1), sorting_eig_pairs[1][1].reshape(784,1))) 
    
    '''Dot product the scaled data with two projections to get close to original data'''
    projecteddata = scaled.dot(stackedcomp)
    projecteddataframe = pd.DataFrame({'projection1': projecteddata[:,0], 'projection2':projecteddata[:,1]})
    
    '''Scatter plot between top two projected components for each class label'''
    print("Scatter Plot between Projected Component 1 and Projected Component 2 for label" + str(i) + ":")
    projecteddataframe.plot(kind="scatter", x= "projection1", y= "projection2", alpha=0.4)
    print("\n")
```

```python
#Solution 6 - (a) (b) (c) (d) (e)

'''
Steps:- 
1. Load the digits dataset. 
2. Create a 30 dimensional projected dataframe.  - Note that similiar steps are followed for PCA as used in solution 5.
3. Consume the projected dataframe from above step for k means clustering.
4. Compute purity and create dataframe with three column -initializaiton method, iteration to converge, purity
'''

# Step 1 & 2 - Load data and get the 30 dimensions projected dataframe that will be used for Clustering -

data = pd.read_csv("train.csv") # import digits data
label_in_data= data['label']

dataval= data.drop(["label"], axis =1)


scaledfeature = StandardScaler(copy=True, with_mean=True).fit_transform(dataval)  # shape is (4684, 784)
cvmat = numpy.cov(scaledfeature, rowvar = False, bias = False )
eigenva, eigenve = numpy.linalg.eig(cvmat) # input to linalg.eig() from numpy module is square array which is covariance matrix in our case
pairs_eigva_eigve = [(numpy.abs(eigenva[k]), eigenve[:,k]) for k in range(len(eigenva))] # create pairs of eigenvalue and eigenvectors
sorting_eig_pairs =sorted(pairs_eigva_eigve, key=lambda eigva: eigva[0], reverse=True) 

stackedcomp = numpy.hstack((sorting_eig_pairs[0][1].reshape(784,1), sorting_eig_pairs[1][1].reshape(784,1),
                           sorting_eig_pairs[2][1].reshape(784,1), sorting_eig_pairs[3][1].reshape(784,1),
                           sorting_eig_pairs[4][1].reshape(784,1), sorting_eig_pairs[5][1].reshape(784,1),
                           sorting_eig_pairs[6][1].reshape(784,1), sorting_eig_pairs[7][1].reshape(784,1),
                           sorting_eig_pairs[8][1].reshape(784,1), sorting_eig_pairs[9][1].reshape(784,1),
                           sorting_eig_pairs[10][1].reshape(784,1), sorting_eig_pairs[11][1].reshape(784,1),
                           sorting_eig_pairs[12][1].reshape(784,1), sorting_eig_pairs[13][1].reshape(784,1),
                           sorting_eig_pairs[14][1].reshape(784,1), sorting_eig_pairs[15][1].reshape(784,1),
                           sorting_eig_pairs[16][1].reshape(784,1), sorting_eig_pairs[17][1].reshape(784,1),
                           sorting_eig_pairs[18][1].reshape(784,1), sorting_eig_pairs[19][1].reshape(784,1),
                            sorting_eig_pairs[20][1].reshape(784,1), sorting_eig_pairs[21][1].reshape(784,1),
                             sorting_eig_pairs[22][1].reshape(784,1), sorting_eig_pairs[23][1].reshape(784,1),
                              sorting_eig_pairs[24][1].reshape(784,1), sorting_eig_pairs[25][1].reshape(784,1),
                              sorting_eig_pairs[26][1].reshape(784,1), sorting_eig_pairs[27][1].reshape(784,1),
                               sorting_eig_pairs[28][1].reshape(784,1), sorting_eig_pairs[29][1].reshape(784,1)))   


projecteddata = scaledfeature.dot(stackedcomp)  # dot product of projections with originial scaled data

colnames = []
for i in range(0,30):
    colnames.append("projection" + str(i))
colnames

projecteddataframe = pd.DataFrame(data = projecteddata, columns = colnames, dtype = None)
DIGITS_PCA_30  = pd.DataFrame(data = projecteddata, columns = colnames, dtype = None)

projecteddataframe['label'] = label_in_data  # append label column to projected data frame
DIGITS_PCA_30['label'] = label_in_data

"""projecteddataframe is the DIGITS_PCA_30 dataframe as mentioned in the question."""
projecteddataframe.head()  # shape is 5 * 31 . since last column is label column 

```

```python
#Solution 6 (a) (b) (c) (d) (e) continued......

# Step 3 - Use the projected 30 dimensions digits projected data for k means clustering- 

"""
Steps -
1.Subset the data for each class. Drop the label column and store values in object 'subsetval'
2.Initialize the 10 cluster centers for each class
3.Generate random centroid points with each point in 30 dimensions. Therefore shape would be 10*30.
    - generate random sample points farther from each other using central limit theoram
    - Farther points -range((mean -4.5*standard deviation), (mean + 5.5*standard deviation))
    - Store initial centroid points in object 'centroidpoints'
    
4. Create a matrix with zero having same shape as of object 'centroidpoints'.
    - This matrix will have the centroid points updated after each iteration
5. Calculate the initial error = distance of random centroid points with origin (0,0,0... )
    - Error will be updated in each iteration -distance of centroid points in iteration (t+1) 
      from centroid points in iteration (t)
6. Calculate the distance of each datapoint(in 30 dimensional space) from each of cluster centroids.
    - By this, we get a list of distances from centroid point corresponding to each datapoint.
    - Sort the list and get the cluster index number corresponding to the minimum distance value.
    - Get all the cluster index values for all data points (in 30 dimensions) in our subsetval data.
7. Filter out the datapoints belonging to a partrituclar cluster.
8. Update the mean of cluster in each iteration. 
9. Repeat step 5 and keep printing error till error becomes zero. Solution converges when error = 0 .
10. Print the error values.
11. Print minimum iterations to converge for each class label.
12. Print the cluster number and count of datapoints in that cluster for each class subset data.
      
"""

cluster_numbers = []  # cluster no will be stored and that will be used in end result dataframe
cluster_points = []  #cluster points(count of datapoints within each cluster are stored for a particular class label)
iter_to_converge = [] # Store iteraation to converge for each class label subset data
lstclass = []
classlabel = []  # we will store the each class label so that in the end, we can build a dataframe having class label and corresponding convergence time
for i in range(0, len(projecteddataframe['label'].unique())):
    classlabel.append(i)
    
    subsetdata =  projecteddataframe[projecteddataframe['label'] ==  projecteddataframe['label'].unique()[i]]
    subsetvar = subsetdata.drop(["label"], axis =1)
    subsetval = subsetvar.values
    subsetval.shape 
    centroidval = [] # create a list of length 30 and each list within list is of length 10.
    for k in range(0,30):
        random.seed(k)
        centroidval.append(random.sample(range(int(numpy.mean(subsetval) - 4.5*numpy.std(subsetval)), int(numpy.mean(subsetval) + 5.5*numpy.std(subsetval))), 10))

    centroidpoints = numpy.transpose(centroidval)  # shape(10*30) since each point is visualized in 30 dimensions/axis.
    centers_initialize = numpy.zeros(centroidpoints.shape, order = "C") 
    #We will store the cluster index value obtained from loop in below section in this variable.
    er = numpy.linalg.norm(centroidpoints - centers_initialize) 
    c = 0
    print("Minimize Error till convergence is reached for class label: " +  str(projecteddataframe['label'].unique()[i]))
    while er!=0:
        
        clusterindex_storage = []
        for rowval in range(0, len(subsetval)):
            distance_with_centroids = numpy.linalg.norm(x = subsetval[rowval] - centroidpoints, keepdims = False,  axis=1)  # shape (10, 0) since we get distance of each row(in 30 dimensions data) with each of the 10 centroid centers.
            arr_to_lst = distance_with_centroids.tolist()
            clusterindex = arr_to_lst.index(min(arr_to_lst))  # Let's say we get clusterindex = 9.
            #That means that the index value associated with minimum number in the list of distances is 9. Note that- list of distances are the distances of each row in our data with centroid centers.
            clusterindex_storage.append(clusterindex)
            #store cluster index of min distance of each row (in our data to be clustered) from centroids centers(shape of centroid center = 10*30 since each centroid center lies in 30 dimensions)
            #shape of clusterindex_storage is (number of rows in data, ) since each value = cluster index corresponding to each observation

        clusterindex_storage = numpy.asarray(clusterindex_storage)

        centers_initialize = deepcopy(centroidpoints)

    #assign number of cluster-
    #df1 = pd.DataFrame()
        for clusterno in range(0, len(label_in_data.unique())):
            rows_in_a_particular_cluster = [subsetval[row] for row in range(len(subsetval)) if clusterindex_storage[row] == clusterno]
    # for example in - [subsetval[row] for row in range(len(subsetval)) if clusterindex_storage[row] == 7], we get only one row 
    #similary there could be possibilities that a set of rows does not belong to any of cluster.
    #fill up centroid centers to which none of the observations belongs to - 
        #(Meaning that, all the observations/rows are far from a particular centroid center). In such case, fill that centroid with previous cluster centroid point.
        #Being the same centroid center, it would not impact the mean of points. Since same values are populated to previous cluster centroids.

            if (len([subsetval[row] for row in range(len(subsetval)) if clusterindex_storage[row] == clusterno]) == 0):
                rows_in_a_particular_cluster = [subsetval[row] for row in range(len(subsetval)) if clusterindex_storage[row] == clusterno-1]
            else:
                rows_in_a_particular_cluster = [subsetval[row] for row in range(len(subsetval)) if clusterindex_storage[row] == clusterno]

            centroidpoints[clusterno] = numpy.mean(rows_in_a_particular_cluster, axis=0)

        er = numpy.linalg.norm(centroidpoints - centers_initialize)
        
        print(er)
        c = c + 1
        
    clustercenters, clusterpoints = numpy.unique(clusterindex_storage, axis = None, return_counts=True, return_inverse=False)
    print("\n")
    print("k means clustering on data subsetted based with class label => " + str(projecteddataframe['label'].unique()[i]))
    print("\n")
    print(pd.DataFrame({'Cluster Number':clustercenters, 'ClusterPoints':clusterpoints, 'Iterations_to_converge': c}))
    list_cc = list(clustercenters)
    list_pp = list(clusterpoints)
    listclass = list(str(projecteddataframe['label'].unique()[i]))
    listclass = numpy.repeat(listclass, len(clustercenters))
    itertoconverge = numpy.repeat(c, len(clustercenters))
    
    cluster_numbers.extend(list_cc)  #Cluster number for a class label to which datapoints belongs or are nearest to.
    cluster_points.extend(list_pp)   # Datapoints associated with each cluster for a particular class label.
    lstclass.extend(listclass)
    iter_to_converge.extend(itertoconverge)
    print("Total iterations to converge:", c)
    print("\n")
    
final_clustering_summary = pd.DataFrame(list(zip(lstclass, cluster_numbers, cluster_points, iter_to_converge)), columns =['Class_Label', 'Cluster_Number', "Datapoints_in_Cluster", "Convergence_Iter_for_classlabel"]) 
print("Final detailed summary after final iterations(i.e. point when no cluster centroid change happens) for each class label:-")
final_clustering_summary
```

```python
#Solution 8
##################

"""
Steps:
1. Try with gaussian_nos = 4 i.e. number of gaussians = 4
2. Plot likelihood values for each iteration
3. Try with gaussian_nos = 10 i.e. number of gaussians = 10 and plot likelihood for each iteration
4. Plot Likelihood value for each gaussian taking maximum log likelihood values.(plot between k and log likelihood)
    - Interpretations
    
"""


iris = pd.read_csv("C://Users//nitesh.jindal//OneDrive - Kantar//ISB//ISB Course Content and Sessions//Term 2//Data Mining//Assignment//iris-species//Iris.csv")
print("iris data: ", iris.head())

iris1 = iris.drop(["Species", "Id"], axis =1)
iris2= iris1.values
print("\n")
print("iris2: ",  iris2[1:10])


''' Introduce a dummy number which will be added to covariance matrix in our code to escape from singularity issue
i.e. in case determinant of matrix becomes zero and its gives error in division by zero, we are adding a very small term
which would not effect our interpretation'''

dummyno = 0.01*np.identity(len(iris2[0]), dtype = float)

''' set iteration = 25 in order to get better log likelihood value
Since in each iteration, likelihood to entire data increases by updating mean, covariance & probability associated
with each gaussian'''
noofiter = 25  

""" 
Initialize  mean, covariance and gauss_prob values i.e.probability assigned to each gaussian and sum(gauss_prob)= 1
"""
np.random.seed(7)
init_mean = np.random.randint(np.array(iris2).min(),np.array(iris2).max(),size=(4,len(iris2[0])),  dtype='l') 
# shape of init_mean = (4*4) since 4 gaussians and each gaussian in 4 dimension. 

"""
we require 4 co variance matrix since we have 4 gaussians. With below code, we get 
4 covariance matrix each one for each gaussian & each with shape 4*4 ( dimensions = 4 )
"""
covariance = np.zeros((4,len(iris2[0]),len(iris2[0]))) 
for i in range(len(covariance)):
    np.fill_diagonal(covariance[i],5)
covariance    
              
LL = []  #log likelihoods
gauss_prob = np.ones(4, order='C', dtype=None)/4  # 0.25 probability to each gaussian. shape(4,).prior probabilities assigned to gaussian. 


for i in range(0,noofiter): # iterate over  the number of iterations = 25 to get better log likelihood value
    datapoint_prob_in_gaussian = np.zeros((len(iris2),len(covariance)))
    covariance[0] = covariance[0] + dummyno  # this step will help in escaping from error raised due to singularity issues in covariance matrix.
    covariance[1] = covariance[1] + dummyno  # this step will help in escaping from error raised due to singularity issues in covariance matrix.
    covariance[2] = covariance[2] + dummyno  # this step will help in escaping from error raised due to singularity issues in covariance matrix.
    covariance[3] = covariance[3] + dummyno  # this step will help in escaping from error raised due to singularity issues in covariance matrix

    #calculate total probability. rowsum of prob = 1.
    tot_prob = np.sum([p*multivariate_normal(mean=m,cov=c).pdf(iris2) for m, p,c in zip(init_mean, gauss_prob, covariance+dummyno)],axis=0)
    
    '''probabilities of each datapoint in each of the gaussian - '''
    prob_g1 = gauss_prob[0]*multivariate_normal(mean= init_mean[0],cov=covariance[0]).pdf(iris2)/tot_prob # gaussian 1 probabilities for x(n)
    prob_g2 = gauss_prob[1]*multivariate_normal(mean= init_mean[1],cov=covariance[1]).pdf(iris2)/tot_prob  # gaussian 2 probabilities for x(n)
    prob_g3 = gauss_prob[2]*multivariate_normal(mean= init_mean[2],cov=covariance[2]).pdf(iris2)/tot_prob  # gaussian 3 probabilities for x(n)
    prob_g4 = gauss_prob[3]*multivariate_normal(mean= init_mean[3],cov=covariance[3]).pdf(iris2)/tot_prob  # gaussian 4 probabilities for x(n)
    
    ''' stack column wise so that each row represents probability of a particular dataoint in all 
        the three gaussians'''
    datapoint_prob_in_gaussian = np.column_stack((np.array(prob_g1), np.array(prob_g2), np.array(prob_g3), np.array(prob_g4))) 
    
    
    
    '''Step2: Maximization Step'''
    init_mean = []
    gauss_prob = []
    covariance = []
    
    
    '''In below for loop, 
     datapoint_prob_in_gaussian[0] is an array of probability of a n dimensional datapoint in all 4 gaussians
     shape is (4,) since each datapoint0 has different probabilities in each of the gaussians '''

    for i in range(len(datapoint_prob_in_gaussian[0])): 
        
        #update mean -
        sum_prob_in_gaussian = np.sum(datapoint_prob_in_gaussian[:,i],axis=0)
        updated_mean = (1/sum_prob_in_gaussian)*np.sum(iris2*datapoint_prob_in_gaussian[:,i].reshape(len(iris2),1),axis=0) 
        init_mean.append(updated_mean)
        
        #update gaussian probability -
        gauss_prob.append(sum_prob_in_gaussian/np.sum(datapoint_prob_in_gaussian)) #gauss_prob = probability assigned to each gaussian and sum(gauss_prob)= 1
        
        #update covariance -
        #covariance.append(((1/sum_prob_in_gaussian)*np.dot((np.array(datapoint_prob_in_gaussian[:,i]).reshape(len(iris2),1)*(iris2-updated_mean)).T,(iris2-updated_mean)))+dummyno)
        covariance1 = ((1/sum_prob_in_gaussian)*np.dot((np.array(datapoint_prob_in_gaussian[:,i]).reshape(len(iris2),1)*(iris2-updated_mean)).T,(iris2-updated_mean)))
        covariance2 = covariance1 + dummyno
        covariance.append(covariance2)
       
    #update Log likelihood function value - 
    LL.append(np.log(np.sum([k*multivariate_normal(init_mean[x],covariance[y]).pdf(iris2) for x,k,y in zip(range(len(init_mean)), gauss_prob, range(len(covariance)))])))
print("Gaussian Mean in last iteration: ", init_mean)
print("\n")
print("Gaussian Covariance in last iteration : ", covariance)
print("\n")
print("Gaussian Probabilities: ", gauss_prob)

'''We observe that gaussian 2th contains the largest number of datapoints since its prob = 0.35 followed
by gaussian 3rd.'''

      
      
```
