```python
pathname = "C://Nitesh//ISB Course Content and Sessions//Term 2//Data Mining//Assignment1//instacart-market-basket-analysis//"


products = pd.read_csv("products.csv")
orders = pd.read_csv("orders.csv")
order_products__train = pd.read_csv("order_products__train.csv")
order_products__prior= pd.read_csv("order_products__prior.csv")
departments= pd.read_csv("departments.csv")
aisles= pd.read_csv("aisles.csv")
sample_submission= pd.read_csv("sample_submission.csv")


order_rlup = pd.merge(order_products__train, orders[['order_id', 'user_id', 'order_dow']],on='order_id', how='left')
product_rlup = pd.merge(order_rlup, products[['product_id', 'product_name']], on='product_id', how='left')
product_lst=product_rlup.groupby(by="user_id", axis=0)['product_name'].apply(list)
product_lst_df= pd.DataFrame(product_lst)
product_lst_df.columns = ["itemsets"]
purchase_freq_wkly= list(product_lst_df['itemsets'])
#weeklyitemspur_by_user = weeklyitemspur_by_user[1:50000]
del orders, order_products__train, products,order_rlup,product_rlup,product_lst,product_lst_df
purchase_freq_wkly


"""
Item Purchase Frequency by Users per week
"""

import pandas as pd
import mlxtend
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
t= TransactionEncoder()
t_array= t.fit(purchase_freq_wkly).transform(purchase_freq_wkly)

transactionfreq = pd.DataFrame(t_array, columns=t.columns_)
transactionfreq

columnlist = []
for i in range(0, 39123):
    test = transactionfreq[transactionfreq.iloc[:,i:i+1].columns[0]].sum()
    if(test > 5 & test < 8):
        columnlist.append(transactionfreq.columns[i])
        
    
new_trans_freq= transactionfreq[columnlist]
new_trans_freq
```

```python
print("Minimum support values for k =5,10,15,20: ")
minimum_support5= 5/new_trans_freq.shape[0]
print(minimum_support5)

'''Item should have appeared atleast 10 times in the transactions. 
For each product, set minimum support with Weekly Transactions greater than 10 is: '''

minimum_support10= 10/new_trans_freq.shape[0]
print(minimum_support10)

minimum_support15=15/new_trans_freq.shape[0]
print(minimum_support15)

minimum_support20= 20/new_trans_freq.shape[0]
print(minimum_support20)

""" Compute dataframe with different support frequenciesi.e. 5, 10, 15, 20
with columns support threshold and product pair. 
"""
support_and_product5 = apriori(new_trans_freq, use_colnames=True,  min_support = minimum_support5)
support_and_product10 = apriori(new_trans_freq, use_colnames=True, min_support = minimum_support10)
support_and_product15 = apriori(new_trans_freq, use_colnames=True, min_support = minimum_support15)
support_and_product20 = apriori(new_trans_freq, use_colnames=True, min_support = minimum_support20)
```
